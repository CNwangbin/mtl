```python
import pandas as pd

df_test = pd.read_pickle('work_dirs/simple/simple_predictions.pkl')
```

```python
species = pd.read_pickle('data/pfam/species.pkl').species.values
```

```python
p_cnt = 0
for i,row in df_test.iterrows():
    if "Bacteria" in row.species:
        print('---'*5)
        print([species[i] for i, val in enumerate(row.labels.squeeze(0)) if val == 1])
        print([species[i] for i, val in enumerate(row.preds) if val > 0.33])
        p_cnt += 1
        if p_cnt > 100:
            break
```
